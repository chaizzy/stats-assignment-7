```python
# ANS 1:
import scipy.stats as stats

sample_mean = 50
sample_std = 5
# assume sample size 30
sample_size = 30 

confidence_level = 0.95
significance_level = 0.05

margin_of_error = stats.norm.ppf(1 - significance_level / 2) * (sample_std / (sample_size ** 0.5))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Confidence Interval:", confidence_interval)
```

```python
# ANS 2
import scipy.stats as stats
import numpy as np
# assuming colors data
# null hypothesis -- data meets the expectation 
# alternate hypothesis --- data does not meet the expection
observed_data = [25 , 30 , 25 , 15, 13, 14]
expected_data = [20, 28, 24, 10, 16, 24]
```

```python
chisquare_test_statistics , p_value = stats.chisquare(observed_data, expected_data)
```

```python
chisquare_test_statistics , p_value
```

```python
significance = 0.05
dof=len(expected_data)-1
print(dof)
critical =stats.chi2.ppf(1-significance,dof)
```

```python
if chisquare_test_statistics > critical:
    print("reject the null hypothesis")
else:
    print("we fail to reject the null hypothesis")
```

```python
# ANS 5 :
import scipy.stats as stats
import math 

# ASSUME sample size 30
sample_mean = 75
sample_std = 12
sample_size = 30
significance_level=0.10
z_score = stats.norm.ppf(1 - (significance_level) / 2)

margin_of_error = z_score * (sample_std / math.sqrt(sample_size))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error


print("sample mean:",sample_mean)
print("Sample standard deviation:",sample_std)
print("90% Confidence interval" , lower_bound ,"-", upper_bound)
 
```

```python
# ANS 6:
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

df = 10  # Degrees of freedom

x = np.linspace(0, 30, 500)  
y = stats.chi2.pdf(x, df)  

plt.plot(x, y, label=f"df = {df}")


x_fill = np.linspace(15, 30, 500)  
y_fill = stats.chi2.pdf(x_fill, df)  
plt.fill_between(x_fill, y_fill, 0, alpha=0.3, color='blue', label="Chi-square statistic = 15")


plt.xlabel("Chi-square statistic")
plt.ylabel("Probability density")
plt.legend()


plt.show()
```

```python
# ANS 9 :
import scipy.stats as stats
import numpy as np
from scipy.stats import chi2_contingency


observed = np.array([[60, 140],[30, 170]])
alpha = 0.05

chi2, p, dof, ex = chi2_contingency(observed, correction=False)
```

```python
chi2,p,ex,dof
```

```python
critical =stats.chi2.ppf(1-alpha,dof)
```

```python
critical
```

```python
if chi2 < critical:
    print("There is no a significant association between smoking status and lung cancer diagnosis.")
else:
    print("There is  significant association between smoking status and lung cancer diagnosis.")
```

```python
# ANS 11:
import scipy.stats as stats

# null_hypothesis = population mean is 70
# alternate_hypothesis = population is significantly different from 70
# it is a two tail test

sample_mean = 72
population_mean = 70
sample_std = 10
sample_size = 30
significance_level = 0.05

# Compute the test statistic
t = (sample_mean - population_mean) / (sample_std / (sample_size ** 0.5))

critical_value = stats.t.ppf(0.9750 , 49)

if (t) > critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

print("Test statistic (t):", t)
print("Critical value:", critical_value)
print("Conclusion:", conclusion)


```
